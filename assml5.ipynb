```python
import pandas as pd
import csv
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
df = pd.read_csv('Market_Basket_Optimisation.csv')
```

```python
df
```

```python
dataset = []
with open('Market_Basket_Optimisation.csv') as file:
    reader = csv.reader(file, delimiter=',')
    for row in reader:
        dataset += [row]
        
```

```python
dataset[1:10]
```

```python
dataset[0:10]
```

```python
len(dataset)
```

```python
te = TransactionEncoder()
x = te.fit_transform(dataset)
```

```python
x
```

```python
te.columns_
```

```python
df = pd.DataFrame(x, columns=te.columns_)
```

```python
len(te.columns_)
```

```python
df.head()
```

```python
freq_itemset = apriori(df, min_support=0.01, use_colnames = True)
```

```python
freq_itemset
```

```python
rules = association_rules(freq_itemset, metric='confidence', min_threshold=0.25)
```

```python
rules
```

```python
rules = rules[['antecedents','consequents','support','confidence']]
```
